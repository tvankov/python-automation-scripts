```python
import os
import csv

gesamt_dicts = {}

start_ordner = os.path.abspath('filialen')
print("Startordner:", start_ordner)

if not os.path.exists(start_ordner):
    print("Ordner existiert nicht!")
else:
    for dirpath, dirnames, filenames in os.walk(start_ordner):

        print("Aktueller Pfad:", dirpath)
        print("Unterordner:", dirnames)
        print("Dateien:", filenames)
        print("-" * 30)

        for filename in filenames:
            print(filename, filename.endswith('.csv'))

            if filename.endswith('.csv'):
                full_path = os.path.join(dirpath, filename)

                with open(full_path, mode='r') as file:
                    reader = csv.DictReader(file)

                    for r in reader:
                        print(r)

                        produkt = r.get('Produkt')
                        tagesumsatz = r.get('Tagesumsatz')

                            
                        if produkt in gesamt_dicts:
                            gesamt_dicts[produkt] += float( tagesumsatz )
                        else:
                            gesamt_dicts[produkt] = float( tagesumsatz )

        print('gesamt_dicts: ', gesamt_dicts)

#################################################################################################
print(os.getcwd())
print()

old_dicts = {}

file_path = os.path.join(os.getcwd(), 'gesamt_information', 'gesamt_information.csv')

with open(file_path, mode='r', encoding='UTF-8', newline='') as file:

    reader = csv.DictReader(file)

    for raw in reader:
        produkt = raw.get('Produkt')
        tagesumsatz = raw.get('Tagesumsatz')
        
        old_dicts[produkt] = float( tagesumsatz )
    
    print(old_dicts)
#################################################################################################

for key, value in gesamt_dicts.items():

    if key in old_dicts:
        old_dicts[key] += value
    else:
        old_dicts[key] = value

print('-'*34)
print(old_dicts)
print('-'*34)

################################################################################################

with open(file_path, mode='w', encoding='UTF-8', newline='') as file:

    spalten = ['Produkt','Tagesumsatz']
    
    writer = csv.DictWriter(file, fieldnames=spalten)

    writer.writeheader()
    writer.writerows([
        {'Produkt': key, 'Tagesumsatz': value}
        for key, value in old_dicts.items()
    ])


    
```
